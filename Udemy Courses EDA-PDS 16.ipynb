### <a id='4a'>4a HYPOTHESIS TEST PART 1</a> 

<a id='4.1'>4.1. What is the association between course duration and the level of course difficulty?</a> 

*Course time duration and level of difficulty.*

Theoretically, a course would be lengthy if it is more engaging or difficult. This would be tested.

*Ho (Null Hypothesis)* : There isnt an association between course duration and the level of courses difficulty so the length in course duration has nothing to do with it.

*Ha (Alternate Hypothesis)* : There is an association between course duration and the level of courses difficulty explaining the length in course duration or vice versa.

**BRIEF INTRO ON PEARSON CHI-SQUARE**

The analysis would be conducted using Pearson chi-square statistic.
Chi Square statistic is commonly used for testing relationships between categorical variables. 

The null hypothesis of the Chi-Square test is that no relationship exists on the categorical variables in the population; they are independent.

The Chi-Square statistic is most commonly used to evaluate Tests of Independence when using a crosstabulation (also known as a bivariate table).
The Test of Independence assesses whether an association exists between the two variables by comparing the observed pattern of responses in the cells to the pattern that would be expected if the variables were truly independent of each other (statitics solution 2020). 

```python
table = pd.crosstab(df['level'],df['content_duration'])
# Get chi-square value , p-value, degrees of freedom, expected frequencies using the function chi2_contingency
stat, p, dof, expected = chi2_contingency(table)
# select significance value
alpha = 0.05
# Determine whether to reject or keep null hypothesis
print('significance=%.3f, p=%.3f' % (alpha, p))
if p <= alpha:
    print('Variables are associated (reject H0)')
else:
    print('Variables are not associated(fail to reject H0)')
```

```python
fig = px.box(
    df,
    x='level',
    y='content_duration',
    color='level',
    title='Boxplot of Level and Course content Duration')
fig.update_yaxes(range=[0,40])
fig.update_layout(showlegend=False)
fig.show()
```

<a id='4.2'>4.2. Summary</a> 

The Pearson chi square test shows that the variables are not associated and for that reason, we fail to reject the null hypothesis because indeed the length of courses are not dependant on the level of difficulty as seen in the boxplot as well.

### <a id='4b'>4b HYPOTHESIS TEST PART 2</a> 

*Course patronage over the years*

This would test to see if there has been a demand in the patronage of Udemy courses and subjects over the past years.

*Ho (Null Hypothesis)* : Udemy hasnt had a substantial patronage in their courses and subjects over the past years.

*Ha (Alternate Hypothesis)* : Udemy has had a substantial patronage in their courses and subjetcs over the past years.

<a id='4.3'>4.3 Has there been a substantial growth in the demand for Udemy courses over the past years?</a> 

```python
#reassigning to data type string
df['course_id'] = df['course_id'].astype(str)

#Converting to pandas datetime
df['published_timestamp'] = pd.to_datetime(df['published_timestamp'])
df['date_published'] = df.loc[:, 'published_timestamp'].apply(lambda s: s.date())
df['year_published'] = df.loc[:, 'published_timestamp'].apply(lambda s: s.year)
df['month_published'] = df.loc[:, 'published_timestamp'].apply(lambda s: s.month_name())

#reset index for visual computation
subject = df['subject'].value_counts().reset_index()
subjects = df['subject'].unique()

subset = df[['date_published','subject']]
subset = subset.sort_values('date_published')
time_series = subset['date_published'].value_counts().reset_index()
time_series.columns = ['Date', 'Counts']
time_series = time_series.sort_values('Date')
time_series['Cum Count'] = time_series['Counts'].cumsum()
dummies = pd.get_dummies(subset['subject'])

subset = subset.join(dummies)
subset['Cum Business'] = subset['Business Finance'].cumsum()
subset['Cum Software'] = subset['Web Development'].cumsum()
subset['Cum Music'] = subset['Musical Instruments'].cumsum()
subset['Cum Design'] = subset['Graphic Design'].cumsum()
subset_melt = subset.melt(id_vars='date_published', value_vars=['Cum Business', 'Cum Software', 'Cum Design', 'Cum Music'])

#plotly scatterplot of courses and courses by subject
fig = make_subplots(
    rows=2, 
    cols=1,
    subplot_titles=("Number of courses",
                    "Number of courses by subject"))
df.sort_values('date_published', inplace=True)
fig.append_trace(go.Scatter(
    x=time_series['Date'],
    y=time_series['Cum Count'],
    name="All",
    mode='lines'),
    row=1, col=1)

fig.append_trace(go.Scatter(
    x=subset['date_published'], 
    y=subset['Cum Business'], 
    mode="lines",
    name="Business",
    line=dict(color="#617C58")
),
    row=2, col=1)
fig.append_trace(go.Scatter(
    x=subset['date_published'], 
    y=subset['Cum Software'], 
    mode="lines",
    name="Software",
    line=dict(color="#74597D", dash="longdashdot"),
),
    row=2, col=1)
fig.append_trace(go.Scatter(
    x=subset['date_published'], 
    y=subset['Cum Design'], 
    
    mode="lines",
    name="Design",
    line=dict(color="#C85A17", dash="dash")
),
    row=2, col=1)
fig.append_trace(go.Scatter(
    x=subset['date_published'], 
    y=subset['Cum Music'], 
    
    mode="lines",
    name="Music",
  
    line=dict(color="#1884C7", dash="dashdot")
),
    row=2, col=1)
fig.update_layout(width=700, height=800)
fig.show()
```

<a id='4.4'>4.4 Summary</a> 

The time series plot shows and evident growth of the rise in demand for Udemy courses particulary from 2014 where it has consistently risen over the past 3 years since 2014.

The time series plot also shows an evident increase in subjects by courses where we see software rise over business particulary in Febuary 2017.

Overall, there has been a substantial growth in the demand for Udemy courses over the past years hence we reject our null hypothesis.

### <a id='4c'>4c HYPOTHESIS TEST PART 3</a> 

*Price and subscribers*

This would test to see if price of courses influences the subscription.

*Ho (Null Hypothesis)* : There is no association between the number of subscribers over the price of the courses

*Ha (Alternate Hypothesis)* : There is an association between the number of subscribers over the price of the courses

<a id='4.5'>4.5 Are people drawn to more expensive courses or not?</a> 

```python
table = pd.crosstab(df['price'],df['num_subscribers'])
# Get chi-square value , p-value, degrees of freedom, expected frequencies using the function chi2_contingency
stat, p, dof, expected = chi2_contingency(table)
# select significance value
alpha = 0.05
# Determine whether to reject or keep null hypothesis
print('significance=%.3f, p=%.3f' % (alpha, p))
if p <= alpha:
    print('Variables are associated (reject H0)')
else:
    print('Variables are not associated(fail to reject H0)')
```

```python
fig = px.violin(
    df,
    x='price',
    y='num_subscribers',
    color='price',
    title='ViolinPlot of Price and Number of subscribers')
fig.show()
```

<a id='4.6'>4.6 Summary</a> 

There is an association between the price of courses and the number of subscribers. It tends out that number of subscribers are in tune more to the free courses hence the higher demand than the paid courses.

The chi-test indicates that the variables price and number of subscribers are associated hence we reject the null hypothesis.
